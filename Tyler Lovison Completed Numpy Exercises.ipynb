## Exercise 4

```python
import numpy as np

arr1d = np.array([10, 20, 30, 40, 50])
print("1-D Array:", arr1d)
```

```python
arr2d = np.array([[10, 20, 30], [40, 50, 60]])
print("2-D Array:\n", arr2d)
```

```python
print("First element of arr1d:", arr1d[0])
print("Second row, third column of arr2d:", arr2d[1,2])
```

```python
print("Slice arr1d from index 1 to 3:", arr1d[1:4])
print("Slice arr2d first row:", arr2d[0, :])
```

```python
num_arr = np.array([30, 20, 40, 10])
print("Sorted numeric array:", np.sort(num_arr))

alpha_arr = np.array(['kiwi', 'berry', 'pear'])
print("Sorted alphabetic array:", np.sort(alpha_arr))
```

```python
zeros = np.zeros(5)
ones = np.ones(5)
range_arr = np.arange(1, 11)
lin_arr = np.linspace(0, 1, 5)

print("Zeros:", zeros)
print("Ones:", ones)
print("Range:", range_arr)
print("Linspace:", lin_arr)
```

## Exercise 5

```python
import numpy as np

arr = np.array([3, 2, 14, 9, 10, 12])

# Find index where value is 9
indexes = np.where(arr == 9)
print("Indexes of 9:", indexes[0])
```
