```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Custom colors for species
species_colors = ['#78C850', '#F08030', '#6890F0']  # Setosa, Versicolor, Virginica

plt.figure(figsize=(10,8))
# Draw violin plot of petal length by species
sns.violinplot(x='species', y='petal_length', data=iris, palette=species_colors)
plt.title("Violin Plot of Petal Length by Species")
plt.show()
```

# Exercise 5 - Draw a Histogram
 Histogram
A histogram is a bar graph-like representation of data that buckets a range of classes into columns along the horizontal x-axis.

1)

sns.histplot(iris, x='petal_length', kde=True)

2) Also try below code with another dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Histogram of petal length for iris
plt.figure(figsize=(8,6))
sns.histplot(data=iris, x='petal_length', kde=True, color='skyblue')
plt.title("Histogram of Petal Length (Iris Dataset)")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Count")
plt.show()


# Load penguins dataset
penguins = sns.load_dataset("penguins")

# Histogram of flipper length for penguins
plt.figure(figsize=(8,6))
sns.histplot(data=penguins, x="flipper_length_mm", kde=True, color='lightgreen')
plt.title("Histogram of Flipper Length (Penguins Dataset)")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Count")
plt.show()

```

# Exercise 6 - Draw a Swarm Plot

A swarm plot is another way of plotting the distribution of an attribute or the joint distribution of a couple of attributes.

Create a Swarm plot of Petal length across the species.
 
sns.swarmplot(y='petal_length', x='species', data=iris, palette=species_colors)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Define custom colors for species
species_colors = {"setosa": "#78C850", "versicolor": "#F08030", "virginica": "#6890F0"}

plt.figure(figsize=(8,6))
sns.swarmplot(x='species', y='petal_length', data=iris, palette=species_colors)
plt.title("Swarm Plot of Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

# Exercise 7 - Draw a Bar Plot

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a column chart.
Execute below command

sns.barplot(penguins, x="island", y="body_mass_g")


```python
# Load penguins dataset
penguins = sns.load_dataset("penguins")

plt.figure(figsize=(8,6))
sns.barplot(x="island", y="body_mass_g", data=penguins, palette="pastel")
plt.title("Bar Plot of Body Mass by Island")
plt.xlabel("Island")
plt.ylabel("Body Mass (g)")
plt.show()
```

# Exercise 8 - Draw a Heatmap Plot

corrmat = iris.corr()

sns.heatmap(corrmat, annot = True, square = True)


```python

import seaborn as sns
import matplotlib.pyplot as plt

#  iris dataset
iris = sns.load_dataset('iris')

#  numeric columns
numeric_iris = iris.select_dtypes(include='number')

#  correlation matrix
corrmat = numeric_iris.corr()

#heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corrmat, annot=True, square=True, cmap="coolwarm")
plt.title("Heatmap of Iris Feature Correlations")
plt.show()

```
