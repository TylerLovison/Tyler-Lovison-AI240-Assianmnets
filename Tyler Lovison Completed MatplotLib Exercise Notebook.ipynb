## Exercise 3 - Draw Barchart




```python
# Data for bar chart
cars = ['Toyota', 'Honda', 'Ford', 'BMW', 'Audi']
horsepower = [130, 150, 180, 210, 190]

# Draw bar chart
plt.bar(cars, horsepower, color='skyblue')
plt.title("Car Horsepower Comparison")
plt.xlabel("Car Brand")
plt.ylabel("Horsepower (HP)")
plt.show()

```

## Exercise 4 -  Draw Histogram


```python
# Example mileage data
mileage = [22, 25, 27, 21, 30, 32, 26, 28, 24, 29, 31, 27, 23, 25, 33]

# Draw histogram
plt.hist(mileage, bins=6, color='lightgrey', edgecolor='Blue')
plt.title("Distribution of Car Mileage")
plt.xlabel("Miles per Gallon (MPG)")
plt.ylabel("Number of Cars")
plt.show()
```

## Exercise 5 -  Draw Piechart



```python
# Data for pie chart
car_types = ['SUV', 'Sedan', 'Truck', 'Coupe']
sales = [40, 35, 15, 10]

# Draw pie chart
plt.pie(sales, labels=car_types, autopct='%1.1f%%', startangle=90, colors=['Green', 'Gold', 'Blue', 'Red'])
plt.title("Car Type Market Share")
plt.show()
```
